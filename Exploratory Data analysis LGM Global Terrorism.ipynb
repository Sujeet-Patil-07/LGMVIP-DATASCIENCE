### Data Visualization


#### 1.Terrorist Activities in Each Year


```python
plt.figure(figsize=(25,12))
sns.countplot(df['Year'], color='red', edgecolor = 'k')
plt.xlabel("Attack year", fontweight = 'bold', fontsize = 18)
plt.ylabel("Number of attacks", fontweight = 'bold', fontsize = 18)
plt.title("Terror Attacks in each year", fontweight = 'bold', fontsize = 25)
plt.xticks(rotation = 90)
plt.show()
```

#### 2. No of success attacks


```python
plt.figure(figsize=(25,12))
sns.countplot(x = df['Year'], hue ='success', data = df, edgecolor = 'k')
plt.xlabel("Attack Year", fontweight = 'bold', fontsize = 18)
plt.ylabel('Number of attacks', fontweight = 'bold', fontsize = 18)
plt.title('Number of Terror Attacks in each year', fontweight = 'bold', fontsize = 25)
plt.xticks(rotation = 90)
plt.legend(title = 'Attacks', loc = 'upper left', labels=['Success Attack', 'Unsuccess Attack'], fontsize= 14)
```

#### 3. Weapons

```python
plt.figure(figsize=(25,12))
sns.countplot(df['Weapon_Type'], order=df['Weapon_Type'].value_counts().index, edgecolor = 'k')
plt.xlabel('Weapon Type',fontweight='bold', fontsize = 18)
plt.ylabel('Number weapons',fontweight='bold', fontsize = 18)
plt.title('Type Of weapons', fontweight='bold',fontsize=25)
plt.xticks(rotation = 20)
```

#### 4. Type of attacks


```python
plt.figure(figsize=(25,12))
sns.countplot(df['Attack_Type'], order=df['Attack_Type'].value_counts().index, edgecolor='k')
plt.xlabel('Attack Type',fontweight='bold', fontsize = 18)
plt.ylabel('Number of Attacks',fontweight='bold', fontsize = 18)
plt.title('Type Of Attacks', fontweight='bold',fontsize=25)
plt.xticks(rotation = 20)
```

#### 5.Type of Target

```python
plt.figure(figsize=(25,12))
sns.countplot(df['Target_Type'], order=df['Target_Type'].value_counts().index, edgecolor='k')
plt.xlabel('Target Type',fontweight='bold', fontsize = 18)
plt.ylabel('Number of Targets',fontweight='bold', fontsize = 18)
plt.title('Type Of Targets', fontweight='bold',fontsize=25)
plt.xticks(rotation = 90)
```

#### 6. Most Active Terrorrist Groups


```python
plt.figure(figsize=(25,12))
sns.barplot(x=df['Group'].value_counts()[1:10].index, y=df['Group'].value_counts()[1:10].values,edgecolor='k')
plt.xlabel('Names of Terrorist Group',fontweight='bold',fontsize=18)
plt.ylabel('Number of Attacks', fontweight='bold',fontsize=18)
plt.title('Most Active Terrorist Groups', fontweight='bold',fontsize=25)
plt.xticks(rotation = 90)
plt.show()
```

#### 7. Top 10 most affected terriost attack regions


```python
plt.figure(figsize=(25,12))
sns.barplot(df['Region'].value_counts()[:10].index, df['Region'].value_counts()[:10].values, edgecolor = 'k')
plt.title('Top 10 most affected terriost attack regions', fontweight = 'bold', fontsize = 25)
plt.xlabel('Region', fontweight = 'bold', fontsize = 18)
plt.ylabel('Counts', fontweight = 'bold', fontsize = 18)
plt.xticks(rotation = 60)
plt.show()
```

#### 8. Top 10 most affected terriost attack countries


```python
plt.figure(figsize=(25,12))
sns.barplot(df['Country'].value_counts()[:10].index, df['Country'].value_counts()[:10].values, edgecolor = 'k')
plt.xlabel('Countries', fontweight = 'bold', fontsize = 18)
plt.ylabel('Counts', fontweight = 'bold', fontsize = 18)
plt.title('Top 10 most affected terriost attack countries', fontweight = 'bold', fontsize = 25)
plt.show()
```

#### 9. Top 10 most affected terriost attack states


```python
plt.figure(figsize=(25,12))
sns.barplot(df['State'].value_counts()[:10].index, df['State'].value_counts()[:10].values, edgecolor = 'k')
plt.xlabel('States', fontweight = 'bold', fontsize = 18)
plt.ylabel('Counts', fontweight = 'bold', fontsize = 18)
plt.title('Top 10 most affected terriost attack States', fontweight = 'bold', fontsize = 25)
plt.show()
```

#### 12. Number of people killed during the terriost attack


```python
plt.figure(figsize=(25,12))
plt.scatter(df['Year'], df['Killed'], color = 'red')
plt.xlabel('Year', fontweight = 'bold', fontsize = 18)
plt.ylabel('Number of people killed', fontweight = 'bold', fontsize = 18)
plt.title("Number of people killed during the terriost attack", fontweight = 'bold', fontsize = 25)
plt.show()
```

#### 13. Number of wounded during the terriost attack

```python
plt.figure(figsize=(25,12))
plt.scatter(df['Year'], df['Wounded'], color = 'b')
plt.xlabel('Year', fontweight = 'bold', fontsize = 18)
plt.ylabel('Number of people Wounded', fontweight = 'bold', fontsize = 18)
plt.title('Number of people wounded during the terriost attack', fontweight = 'bold', fontsize = 25)
plt.show()
```
